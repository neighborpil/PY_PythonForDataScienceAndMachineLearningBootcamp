```python
%matplotlib inline 
# mat plot에서의 그래프를 주피터 노트북에서 직접 볼 수 있도록 해준다
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x, y)
plt.show() 
# plt.show()를 해주면 옆에 Out으로 표시가 된다.
# 주피터 노트북에서는 위에 %matplotlib inline으로 생성하면 plt.show()를 안해줘도 되지만,
# 다른 환경에서는 show()를 해줘야 그래프가 나타난다
```

```python
# 색깔 넣기
plt.plot(x, y, 'r-')
```

```python
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# OBJECT ORIENTED
fig = plt.figure()
# left axis, bottom of the axis, width, height
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title('LARGER PLOT')

axes2.plot(y, x)
axes2.set_title('SMALLER PLOT')
```

```python
# diff between plot and subplot
'''
plot: single plot
subplots: mulitple plots by rows and columes
'''
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

# axes.plot(x, y)
# plt.tight_layout()

'''
for current_ax in axes:
    current_ax.plot(x, y)
'''

axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

```python
# fig = plt.figure(figsize=(3, 2), dpi=100)
# dpi: dops per inch

fig = plt.figure(figsize=(8, 2))

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2))

axes[0].plot(x, y)
axes[1].plot(x, y)
plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend(loc=0)
```
